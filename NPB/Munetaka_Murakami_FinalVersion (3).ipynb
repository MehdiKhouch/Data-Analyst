<b> Glossary : </b>


<b> "OBP" </b> stands for <b> On-base Percentage in baseball </b>. It is a statistic <b> that measures the frequency at which a batter reaches base safely in relation to their number of plate appearances </b>. On-base percentage provides a comprehensive measure of a batter's <b> ability to avoid making outs and get on base through hits, walks, and hit by pitches </b>. It reflects how effectively a batter contributes to their team's offensive performance by extending innings and creating scoring opportunities.

<b> "OPS" </b> stands for <b> On-base Plus Slugging in baseball</b>. It is a comprehensive statistic that combines a player's on-base percentage (OBP) and slugging percentage (SLG) into a single metric. OPS provides a more complete picture of a player's offensive production by incorporating both <b> their ability to get on base (OBP) and their ability to hit for power (SLG)</b>. It evaluates a player's performance in two critical aspects of hitting. OPS correlates strongly with a team's ability to score runs and win games. <b> Players with higher OPS tend to contribute more to their team's offensive success </b>by generating scoring opportunities and driving in runs.

<b> "WAR" </b> stands for <b> Wins Above Replacement in baseball</b>. It is a comprehensive statistic that <b> quantifies a player's total contribution to their team's success compared to a replacement-level player</b>. WAR combines a player's offensive, defensive, and baserunning contributions into a single value. WAR allows for <b> easy comparison of players </b> across different positions and skill sets. <b> WAR correlates strongly with a team's ability to win games </b>and contend for championships.

### Chapter 2 : Murakami's plate discipline

In Chapter 1, we noted that Murakami had achieved a historic season with the bat in 2022. \
Since then, however, he has come back to very good standards, but below his previous seasons. Soon, we could see that the problem might be his declining ability to make regular contact with the bat.

In this chapter, we will focus on Murakami's ability to make contact with the ball and his batting average. Understand why he declined in 2023 and whether 2022 may not have been an exception.

To evaluate the plate discipline of Murakami, we will use 3 differents angles : his approach at the plate, vision and contact.

Firstly, we will begin with <b>his approach at the plate</b>.\
Behind this confusing term, we will try to understand how Murakami choose his swing and what's his style facing the pitcher.\
Is he more patient, waiting for the right ball to strike, or is he more aggressive? Does he go for the contact or the big hit?

For that, we will use two different metrics : BB% and K%

<b>Glossary :</b>

<b>BB%</b> stands for <b>Walk Percentage</b> in baseball. It is a statistic that measures the <b>frequency at which a batter draws a walk</b>, also known as a base on balls, in relation to their number of plate appearances.\ Walks are valuable for a team because they put a runner on base without the need for a hit. <b>A high BB% indicates that the batter has good patience and discipline at the plate</b>, as they are able to lay off pitches outside the strike zone and draw walks to get on base.

This statistic reflects a batter's ability to work the count, recognize pitches, and contribute to their team's offensive performance by extending innings and creating scoring opportunities without the need for a hit
    
<b>K%</b> stands for <b>Strikeout Percentage</b> in baseball. It is a statistic that measures the <b>frequency at which a batter strikes out, meaning they are retired by the pitcher without putting the ball in play</b>, in relation to their number of plate appearances.\ While strikeouts are not ideal for a batter, they are sometimes accepted as part of a hitter's profile if they come with other valuable skills, such as power hitting or a high walk rate. However, a high K% indicates that the batter struggles to make contact with the ball, either due to difficulty in making solid contact or swinging and missing at pitches.

This statistic reflects a <b>batter's ability to make contact with the ball and avoid striking out</b>, which can affect their ability to contribute to their team's offensive performance by putting the ball in play and advancing baserunners.

Below, we will refine these first two hypotheses by taking our analysis of Murakami's Plate Discipline a step further.

<b>Plate Discipline statistics</b> tell us how often a hitter swings and makes contact with certain kinds of pitches or how often a pitcher induces swings or contact on certain kinds of pitches. These numbers are very useful for determining the type of hitter or pitcher at which you’re looking.

In the second part of this chapter, we'll look at how Murakami decides to swing and whether he gets more balls in or out of the zone. To do this, we'll use 7 statistics that give us a more detailed insight into the player's habits at the plate.

<b> O-Swing%</b> (Swing Percentage on Pitches Outside the Strike Zone) : It measures the proportion of times a player swings at pitches outside the strike zone, compared to the total number of pitches outside the strike zone he receives.

<b>Z-Swing%</b> (Swing Percentage on Pitches Inside the Strike Zone) : It measures the proportion of times a player swings at pitches inside the strike zone, compared to the total number of pitches in the strike zone he receives.

<b>Swing%</b> (Overall Swing Percentage) :  It represents the percentage of all pitches a player chooses to hit, whether inside or outside the strike zone.

<b>O-Contact% </b> (Contact Percentage on Pitches Outside the Strike Zone) : It measures the proportion of times a player makes contact with pitches outside the strike zone, compared to the total number of pitches outside the strike zone he receives.

<b>Z-Contact%  </b> (Contact Percentage on Pitches Inside the Strike Zone) : It measures the proportion of times a player makes contact with pitches inside the strike zone, compared to the total number of pitches in the strike zone he receives.

<b>Contact% </b> (Overall Contact Percentage) : It represents the percentage of all pitches a player chooses to hit that make contact, whether inside or outside the strike zone.

<b>Zone% </b> (Zone Percentage) : It measures the proportion of throws a player receives that are thrown into the strike zone, in relation to the total number of throws he receives.

Murakami is recognized for his wonderful performance in batting, but not for his fielding quality.

In this chapter, we will try to highglight the fielding performances of Murakami thanks to 4 different statistics.


<b> "FP" </b> stands for <b> Fielding Percentage in baseball </b>. It is a statistic <b> that measures the proportion of defensive chances successfully converted into outs by a fielder </b>. Fielding percentage provides a simple measure of a fielder's <b> reliability in making routine defensive plays </b>. It reflects the percentage of times a fielder successfully completes defensive plays without committing errors.


<b> "E" </b> stands for <b> Errors in baseball </b>. It is a statistic <b> that measures the number of defensive mistakes or errors committed by a fielder </b>. Errors occur when a fielder fails to make a play that should have been made with ordinary effort, resulting in the advancement or reaching base of an opposing batter or baserunner.


<b> "RF" </b> stands for <b> Range Factor in baseball </b>. It is a statistic <b> that measures the range or extent of a fielder's defensive coverage on the field </b>. Range factor calculates the average number of defensive plays made by a fielder per game or inning. A higher range factor indicates a fielder who covers more ground and has a wider defensive reach on the field.


<b> "UZR" </b> stands for <b> Ultimate Zone Rating in baseball </b>. It is a statistic <b> that measures the number of runs above or below average a fielder contributes defensively, based on their ability to make plays within their defensive zone </b>. Ultimate zone rating takes into account factors such as a fielder's range, arm strength, and ability to convert batted balls into outs, providing a comprehensive measure of their overall defensive value. A positive UZR indicates above-average defensive performance, while a negative UZR suggests below-average performance.
