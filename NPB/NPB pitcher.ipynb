```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel(r"C:\Users\Mehdi\Desktop\LeaderPitchingNPB2022.xlsx")
```

```python
df_sortedERA = df.sort_values(by="ERA", ascending = True).head(10)
df_sortedK = df.sort_values(by="K%", ascending = False).head(10)
df_sortedWAR = df.sort_values(by="WAR", ascending = False).head(10)
```

```python
df_player = pd.concat([df_sortedERA, df_sortedK, df_sortedWAR]).drop_duplicates()
```

```python
df_players = df_player.sort_values(by = "WAR", ascending = False)
```

```python
sns.scatterplot(data=df_players, x="WAR", y="ERA", hue = "K%")
plt.gcf().set_size_inches(8, 6)

for index, row in df_players.iterrows():
    plt.text(row['WAR'], row['ERA'], row['Player'], fontsize=9, ha='center', va='top')
    
ax = plt.gca()
ax.set_xlim(left=0, right=7)
ax.set_ylim(bottom=0, top= 4)
plt.show()
```

```python
sns.scatterplot(data=df, x="WAR", y="ERA", hue = "K%")
for index, row in df_sortedWAR.iterrows():
    plt.text(row['WAR'], row['ERA'], row['Player'], fontsize=9, ha='center',va='top')
plt.gcf().set_size_inches(8, 6)
ax = plt.gca()
ax.set_xlim(left=0, right=7)
ax.set_ylim(bottom=0, top= 5)
plt.show()
```
